Basics
- Create arrays: x, y  and z where:
    - x contains 12 evenly spaced numbers between 1 and 20 
    - y contains the values of x multiplied by 2 ie. 2*x 
    - z contains the squares of elements of x ie x**2

- Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’. Increase the x axis range to (0, 25) and y axis range to (0, 450)

- Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two  functions using legend.

- Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.

- Finally, redo all the above plots using seaborn which is a higher level plotting library. It creates more aesthetically pleasing charts than matplotlib.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### x contains 12 evenly spaced numbers between 1 and 20

```python
x = np.linspace(1,20,12)
x
```

#### y contains the values of x multiplied by 2 ie. 2*x

```python
y = 2 * x
y
```

#### z contains the squares of elements of x ie x**2

```python
z = x ** 2
z
```

### Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’.

```python
plt.plot(x,y)
plt.ylabel('y')
plt.xlabel('x')
plt.title('y=2x')
plt.grid()
plt.show()
```

###  Increase the x axis range to (0, 25) and y axis range to (0, 450)

```python
plt.plot(x,y)
plt.ylabel('y')
plt.xlabel('x')
plt.title('y=2x')
plt.xlim((0,25))
plt.ylim((0,450))
plt.grid()
plt.show()
```

## Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two functions using legend.

```python
plt.plot(y,x,'b--',label="*2")
plt.plot(z,x,'r--',label="^2")
plt.grid()
plt.legend()
plt.show()
```

## Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.

```python
plt.figure(num=1,figsize=(12,6))
plt.subplot(211)
plt.xlabel('y')
plt.ylabel('x')
plt.title('Y vs X')
plt.grid()
plt.plot(y,x,'b--',label="*2")
plt.subplot(212)
plt.ylabel('y')
plt.xlabel('z')
plt.title('Z vs Y')
plt.grid()
plt.plot(z,x,'r--',label="^2")
plt.subplots_adjust(hspace=0.4)
plt.show()
```

### Finally, redo all the above plots using seaborn which is a higher level plotting library. It creates more aesthetically pleasing charts than matplotlib.

```python
sns.set_style("whitegrid")
sns\
    .lineplot(x=x,y=y)\
    .set(xlabel='x', ylabel='y',title='y=2x')
plt.show()
```

```python
sns.set_style("whitegrid")
sns\
    .lineplot(x=x,y=y)\
    .set(xlabel='x', ylabel='y',title='y=2x',xlim=(0, 25), ylim=(0, 450))
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x=z,y=x,label="*2").lines[0].set_linestyle("--")
sns.lineplot(x=y,y=x,label="^2",color='red').lines[0].set_linestyle("--")
plt.show()
```

```python
dataset = pd.DataFrame(np.array([x,y,z]).T,columns="x y z".split())
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 6))
sns.lineplot(x='y',y='x',data=dataset,ax=axes[0]).lines[0].set_linestyle("--")

sns.lineplot(x='z',y='x',data=dataset,ax=axes[1],color='red').lines[0].set_linestyle("--")
```

###  Create 3 arrays:

```python
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
x2 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
y = np.array([0, 1, 1, 0, 0, 1, 0, 0, 1, 0, 0, 1, 1])

df = pd.DataFrame(np.array([x1,x2,y]).T,columns="x1 x2 y".split())
df
```

### Plot the scatterplot of x1 and x2. Color the points in the scatterplot according to their values in y: blue for 0 and red for 1. For example, color the point (5, 99) blue as the corresponding y value is 0 and the point (7, 86) red as the corresponding y value is 1.

```python
sns.set_style('whitegrid')
sns\
    .FacetGrid(df,hue="y",height=6,palette=['blue','red'],xlim=(df['x1'].min()-5,df['x1'].max()+5),ylim=(df['x2'].min()-5,df['x2'].max()+5))\
    .map(plt.scatter,"x1","x2")\
    .add_legend()
plt.show()
```

The value in class label '0' is double the value in class label '1'. This is case of imbalanced dataset. When we have imbalanced dataset, it will have an bias on the model. 

### Create a figure containing 8 subplots arranged in (2, 4) grid. Plot the histograms of the 8 features (all the columns except the outcome) present in the dataset. Identify the features having normal distribution and skewed distributions.

```python
fig, axes = plt.subplots(2, 4, figsize=(18, 10))
fig.suptitle('Pima Dataset')


for i in range(0,2):
    for j in range(0,4):
        sns.histplot(diabetic_dataset[diabetic_dataset.columns[i*4+j]],ax=axes[i, j],kde=True)
```

### Create a figure containing 8 subplots arranged in (4, 2) grid. Plot the boxplot for each of the features in each of the subplots. Outliers are a set of observations that vary significantly from the other observations. Which feature do you think has the most outliers? And Why?

```python
fig, axes = plt.subplots(4, 2, figsize=(18, 25))
fig.suptitle('Pima Dataset')
for i in range(0,4):
    for j in range(0,2):
        sns.boxplot(x='Outcome',y=diabetic_dataset[diabetic_dataset.columns[2*i+j]],ax=axes[i, j],data=diabetic_dataset,)
```

It looks columns like 'Glucose', 'Insulin', 'DiabeticPedigreeFunction' and 'Age' have outliers comparing them with the median value.
In my opinion, 'Insulin' has the highest number of outliers because so many of their are points are outside of whiskers of the plot in both 'Outcomes'.

### A correlation matrix represents how much the features are correlated to each other. You can compute the correlation matrix of a dataframe using the dataframe.corr() method.
```python
corr = diabetic_dataset.corr()
corr
```
