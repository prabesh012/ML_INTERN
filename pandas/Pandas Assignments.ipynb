Create two dataframe (first dataframe having 50 random rows and second dataframe having 40 random rows). Each dataframe should have 3 columns including a common column (example: id) and 36 rows of data in the second dataframe should be identical to the first dataframe.
- Merge both the dataframe using:
    - Inner
    - Outer
    - Left
    - Right
    - Evaluate the differences observed 
- Use join() to combine the dataframe
- Use concat to combine the dataframe
- Evaluate the differences observed between merge, join and concat



## Create two dataframe (first dataframe having 50 random rows and second dataframe having 40 random rows). Each dataframe should have 3 columns including a common column (example: id) and 36 rows of data in the second dataframe should be identical to the first dataframe.


```python
first_arr = np.concatenate((np.arange(50)[:,np.newaxis],randint(1,100,size=(50,2))),axis=1)
first_arr
```

```python
common_arr = np.concatenate((np.arange(36)[:,np.newaxis],randint(1,100,size=(36,2))),axis=1)
uncommon_arr = np.concatenate((np.array([111,222,333,444])[:,np.newaxis],randint(1,100,size=(4,2))),axis=1)
```

```python
second_arr = np.concatenate((common_arr,uncommon_arr),axis=0)
second_arr
```

```python
df_first = pd.DataFrame(first_arr,columns='ID A B'.split())
```

```python
df_second = pd.DataFrame(second_arr,columns='ID C D'.split())
```

```python
df_first.head()
```

```python
df_second.head()
```

## Merge both the dataframe using

## INNER

```python
pd.merge(df_first,df_second,how='inner',on='ID')
```

## OUTER

```python
pd.merge(df_first,df_second,how='outer',on='ID')
```

## LEFT

```python
pd.merge(df_first,df_second,how='left',on='ID')
```

## LEFT AGAIN but reversing the dataframe

```python
pd.merge(df_second,df_first,how='left',on='ID')
```

## RIGHT

```python
pd.merge(df_first,df_second,how='right',on='ID')
```

## Retrieve the active cases of “Madhya Pradesh” Province from df_india.

```python
df_india[df_india['Province_State'] == 'Madhya Pradesh'][['Active']]
```

## Retrieve the state with the maximum number of confirmed cases from df_india.

```python
df_india[df_india['Confirmed'] == df_india['Confirmed'].max()][['Province_State']]
```

## Filter the Country_Region whose value == “Canada” and assign it in dataframe df_canada.

```python
df_canada =  df_covid[df_covid['Country_Region']=='Canada']
```

```python
df_canada.head()
```

## Province/State Diamond Princess  has the “NaN” value for Active number of cases. Fill it with the median of the Active cases.

```python
df_canada[['Province_State','Active']][df_canada['Province_State'] == 'Diamond Princess']
```

```python
median_val = df_canada['Active'].median()
median_val
```

```python
row_index = df_canada[['Province_State','Active']][df_canada['Province_State'] == 'Diamond Princess'].index
```

```python
df_canada.loc[row_index,'Active'] = median_val
```

```python
df_canada[['Province_State','Active']][df_canada['Province_State'] == 'Diamond Princess']
```

## Filter the Country_Region whose value == “US” and assign it in dataframe df_us.

```python
df_us_states = df_us.groupby('Province_State').max()
df_us_states
```

## Re-index the dataframe df_us_states.

```python
df_us_states[df_us_states['Province_State'].isin(us_states['state'])].index
```
